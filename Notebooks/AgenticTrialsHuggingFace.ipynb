```python
import os
from openai import OpenAI

client = OpenAI(
    base_url="https://router.huggingface.co/v1",
    api_key="",
)

completion = client.chat.completions.create(
    model="meta-llama/Llama-3.1-8B-Instruct:scaleway",
    messages=[
        {
            "role": "usuario",
            "content": "dame una lista de los mejores sitios web para buscar \
                        informacion del mercado inmobiliario en medellin, colombia. \
                        la lista debe estar en formatio Json. Cada sitio web debe tener \
                        los siguientes campos: nombre, url, descripcion.",
        }
    ],
)

print(completion.choices[0].message)
```
