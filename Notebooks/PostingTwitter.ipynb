```python
import tweepy
import os
import sys
# Import the shared post_tweet implementation (moved to twitternews package)
sys.path.insert(0, os.path.abspath('..'))

from twitternews.twitter_utils import post_tweet

# Load config from TOML (prefer file, fall back to env vars)
CONFIG_PATH = '../config.toml'

try:
    # Python 3.11+ stdlib
    import tomllib
    def _load_toml(path):
        with open(path, 'rb') as f:
            return tomllib.load(f)
except Exception:
    # pip package fallback
    try:
        import toml
        def _load_toml(path):
            with open(path, 'r', encoding='utf-8') as f:
                return toml.load(f)
    except Exception:
        def _load_toml(path):
            return {}

config = {}
if os.path.exists(CONFIG_PATH):
    try:
        config = _load_toml(CONFIG_PATH) or {}
    except Exception as e:
        print(f"⚠️ Failed to parse TOML config: {e}")

def _get(key, default=None):
    # Prefer config.toml then environment variables
    if isinstance(config, dict) and key in config:
        return config[key]
    return os.environ.get(key, default)

# Read credentials (keys expected in config.toml: TWITTER_API_KEY, TWITTER_API_SECRET,
# TWITTER_ACCESS_TOKEN, TWITTER_ACCESS_TOKEN_SECRET). Fallback to env vars if missing.
TWITTER_API_KEY = _get('TWITTER_API_KEY')
TWITTER_API_SECRET = _get('TWITTER_API_SECRET')
TWITTER_ACCESS_TOKEN = _get('TWITTER_ACCESS_TOKEN')
TWITTER_ACCESS_TOKEN_SECRET = _get('TWITTER_ACCESS_TOKEN_SECRET')

# Export to environment so twitter_utils (and other modules) can pick them up
for k, v in {
    'TWITTER_API_KEY': TWITTER_API_KEY,
    'TWITTER_API_SECRET': TWITTER_API_SECRET,
    'TWITTER_ACCESS_TOKEN': TWITTER_ACCESS_TOKEN,
    'TWITTER_ACCESS_TOKEN_SECRET': TWITTER_ACCESS_TOKEN_SECRET,
}.items():
    if v:
        os.environ.setdefault(k, v)


# The text you want to post
tweet_text = "Hello, Twitter! I am posting this from a Python script using the Tweepy library (API v2). #PythonProgramming"

# Run the function (when running this notebook directly)
if __name__ == '__main__':
    post_tweet(tweet_text)
```
